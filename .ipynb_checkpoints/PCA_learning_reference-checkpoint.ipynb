Correlation & VIF values was used to determine the correlation between feature & target. Also, the MultiCollinearity within the dataset was determined. It is because the high correlation between feature will also inhibit the regression performance. Therefore, we need to select some important features and make sure all the features selected with low VIF.

```python
# show all correlation towards the target
df_corr = df.corr()['diagnosis'].abs().sort_values(ascending=False)
df_corr
```

There is many features and we aimed to target some features with high correlation.

```python
# select the feature with at least 0.6 in the correlation
features = df_corr[df_corr > 0.6].index.to_list()[1:]
features
```

It is still too much and we can try to calculate the multi-collineraity to eliminate some as well.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

def calculate_vif(df, features):
    # create dict for vif & tolerance container
    vif, tolerance = {}, {}
    
    # calculate the tolerance & VIF between features
    for feature in features:
        X = [f for f in features if f != feature]
        X,y = df[X], df[feature]
        
        r2 = LinearRegression().fit(X,y).score(X,y)
        
        tolerance[feature] = 1 - r2
        
        vif[feature] = 1/(tolerance[feature])
    return pd.DataFrame({'VIF': vif, "Tolerance": tolerance})

calculate_vif(df,features)
```

```python
# try to reduce those feature with high VIF and ensure that remaining feature
# with VIF less than 5
# Let select the 4 features that with lowest VIF
features = [
    'worst concave points',
    'mean concave points',
    'mean concavity',
    'worst concavity'
]

calculate_vif(df,features)
```

It seems like it wont work since the VIF value still remain quite high. Let's try with the features proposed in the previous part. Also, we can try to plot the scatter plot for these feature to check with the correlation as well.

```python
# quick plot by using the pairplot from seaborn
import seaborn as sns

features = df_corr[df_corr > 0.6].index.to_list()[1:]
sns.pairplot(df[features])
```

We can try to pick with less linear relationship with each other and try to minimize the VIF value

```python
# after few trials and there is the best VIF values that we obtained
features = [
    'mean radius',
    'worst concave points',
    'mean concavity',
    'worst concavity'
]

calculate_vif(df,features)
```

By looking at the multi-collinearity & correlation, the accuracy was decreased as we are looking at the single varible and some information will be missed by elimination of those features. Therefore, we will try to use PCA to reduce the dimension and check that what can we do to improve the model performance.

Before we apply PCA on the data set, we need to perform a standard scaling on those features.

```python
from sklearn.preprocessing import StandardScaler

# seperate the features & target
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# create the standard sclaer model. fit & transform the data into model
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# create a PCA model
components = None
pca = PCA(n_components = components)

# fit the data into the model
pca.fit(X_scaled)
```

Parameter n_components:

(1) integer : how many principal components you want to reduce

(2) floating number (0 < n < 1) : return number of components needed to capture the specific percentage of variability in the data

(3) None: return the components sames as the number of original features in the data set

```python
# explained variances show the variability
print(pca.explained_variance_ratio_ * 100)
```

First component alone captures about 44% variability in all the features

Second will be around 19% variability

All 30 components got total 100% variability in the data

```python
# cumulative variances 
print(pca.explained_variance_ratio_.cumsum() * 100)
```

We can see that total 100% explained variance

Let's plot a scree plot for the cumsum

(It's great to visualize the cumsum & check with the change)

```python
# plot a scree plot
components = len(pca.explained_variance_ratio_) if components is None else components

plt.plot(range(1, components + 1), np.cumsum(pca.explained_variance_ratio_ * 100))
plt.xlabel("Number of components")
plt.ylabel("Explained variance (%)")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 0.85)
pca.fit(X_scaled)

print("Cumulative Variances (Percentage):")
print(np.cumsum(pca.explained_variance_ratio_ * 100))

components = len(pca.explained_variance_ratio_)
print(f'Number of components: {components}')


plt.plot(range(1, components + 1), np.cumsum(pca.explained_variance_ratio_ * 100))
plt.xlabel("Number of components")
plt.ylabel("Explained variance (%)")
```

For the graph, we can design the components we choose for further investigation by checking the 'elbow'. It means the big drop for the explained variance and those components with small explained variance that can be ignored.

```python
pca_components = abs(pca.components_)
pcs = np.array(pca_components)

df_pc = pd.DataFrame(pcs, columns = df.columns[1:])
df_pc.index = [f'Component {c}' for c in range(1,7)]
df_pc
```

```python
print('Top 4 most important features in each component:')
for row in range(pca_components.shape[0]):
    temp = np.argpartition(-(pca_components[row]),4)
    
    indices = temp[np.argsort((-pca_components[row])[temp])][:4]
    
    print(f'Component {row}: {df.columns[indices].to_list()}')
```

This is the way that we determine the most important features in each component or we can determine the coeffcient of each features in those components.

Let's transform the standardized data of the 30 columns in the dataset to thr 6 principal components:

```python
X_pca = pca.transform(X_scaled)
print(X_pca.shape)
print(X_pca)
```

From the notebook, he used pipeline to work on. It is also a great experience to learn how to use pipeline to handle the data. Let's try to work on it.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

_sc = StandardScaler()
_pca = PCA(n_components = components)
_model = LogisticRegression()

log_regress_model = Pipeline([
    ('std_scaler', _sc),
    ('pca', _pca),
    ('regressor', _model)
])

# Pipeline([('model 1 name', model1),('model 2 name', model2),(etc....)])
```

```python
# perform a split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = True, random_state = random_state)

# train the model pipeline with the train data 
# (Using the original data since the pipeline include standard scaler, PCA & logistic regression model)
log_regress_model.fit(X_train, y_train)

log_regress_model.score(X_test, y_test)
```

After using the PCA, the model accuracy was imporved from 92% to 97%.
