Let's begin to follow Wei-Meng Lee coding and learn some idea. Then, I will try to apply it to other data set.

```python
# import the data set from sklearn
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
# check the number of feature for the data set
print(len(breast_cancer.feature_names))
```

There is 30 features for this data set. It's great to work with the json/ dictionary format instead of CSV/ Dataframe in pandas.

Let's work for the target and distribution of the target.

```python
import numpy as np

print(breast_cancer.target)
print(breast_cancer.target_names)
print(np.array(np.unique(breast_cancer.target, return_counts = True)))
```

In target, value of 0 means malignant tumour and value of 1 means benign tumour. Also, the sample size is not balanced but it's not serious.

```python
import matplotlib.pyplot as plt

# 6 rows & 5 columns
_, axes = plt.subplots(6, 5, figsize=(15,15))

# seperate the data set according to the target tumour
malignant = breast_cancer.data[breast_cancer.target == 0]
benign = breast_cancer.data[breast_cancer.target == 1]

# flatten the 2D array
ax = axes.ravel()

# work for 30 features
for i in range(30):
    bins = 40    # generate 40 equal width bins
    
    # plot the histogram for each feature
    # generate the plot for two different feature
    # plot for target with i feature, red color & 0.5 transperacy (malignant)
    ax[i].hist(malignant[:, i], bins = bins, color = 'r', alpha = .5)
    # plot for target with i feature, blue color & 0.5 transperacy (benign)
    ax[i].hist(benign[:, 1], bins = bins, color = 'b', alpha = .5)
    
    # set the title
    ax[i].set_title(breast_cancer.feature_names[i], fontsize = 12)
    
    # display the legend
    ax[i].legend(['malignant', 'benign'], loc = 'best', fontsize = 8)
    
plt.tight_layout()
plt.show()
```

It's a great to take a look for the data set. If two histogram are separated, this means that the feature is important and it affect the target. For mean radius, the histogram is separated. It means that the bigger size of tumour will be malignant. On the other hand, the overlapped histogram doesn't tell anything about the effect of feature towards the tumour size.
